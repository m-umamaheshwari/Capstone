### DATA PRE-PROCESSING

```python
# max null value rows in both the dataframes needs to be identified.
null_dict={}
for i in range(data_before.shape[0]):
    null_dict[i]=data_before.iloc[i].isnull().sum()

# identifying individuals who have not responded for more than 7 questions 
max_null=[]
for key in null_dict:
    if null_dict[key]>7:
        max_null.append(key)
```

```python

null_dict_aft={}
for i in range(data_present.shape[0]):
    null_dict_aft[i]=data_present.iloc[i].isnull().sum()

# identifying individuals who have not responded for more than 7 questions 
max_null_aft=[]
for key in null_dict_aft:
    if null_dict_aft[key]>7:
        max_null_aft.append(key)
```

```python
max_null.extend(max_null_aft)

#Common+unique index numbers of null values identified
#Drop these rows to maintain data of same individuals among both datasets

null_rows=set(max_null)
```

```python
#define a function for encoding demographics

def encoding_demograph(dataframe):
    
    dataframe['Age'].replace({'20-25':0,'25-30':1,'30-35':2,'35-40':3,'40-55':4,'55+':5},inplace=True)
            
    dataframe['Income'].replace({"I don't earn right now (maybe in future)":0,
                               'Under 3 lakh per annum':1,
                               '3 - 5 lakh per annum':2,
                               '5 - 10 lakh per annum':3,
                               '10 - 20 lakh per annum':4,
                               '20+ lakh per annum':5,
                               },inplace=True)
            
    dataframe['Gender'].replace({'Male' : 0,'Female': 1, 'Prefer not to say' : 2
                                       },inplace=True)
            
    dataframe['Marital_status'].replace({'No' : 0,'Yes': 1
                                       },inplace=True)
            
    dataframe['Loc'].replace(dict.fromkeys(['Telangana','Mars','Karnataka','Andhra Pradesh','Tamil nadu','Hyderabad','TELANGANA',
                             'Telengana','Telangana ','Madhupur (Telangana)','HYD','Andhra pradesh',
                             'Puppalaguda','Andhra Pradesh ','visakhapatnam','Vishakapatnam','Bangalore',
                             'Tamilnadu','Tamil Nadu','Chennai'], 'South India'),inplace=True)
            
    dataframe['Loc'].replace(dict.fromkeys(['Punjab','Chandigarh','Jharkhand','Madhya Pradesh','West Bengal',
                                          'Maharashtra','Chattisgarh','Uttar Pradesh','Odisha','Haryana',
                                          'Delhi','Gujarat','Gujarat ','Rajasthan','Bihar','Sikkim','Up','Shahdol',
                                          'Mp','West Bengal ','Himachal Pradesh','Maharashtra ',
                                          'UP','Bhopal','Chhattisgarh','Pune','Bhilai','CG','New delhi'],
                                         'North India'),inplace=True)
            
    dataframe['Loc'].replace(dict.fromkeys(['paris','USA','Texas','Hertfordshire','sydney','Melbourne','Pennsylvania',
                            'singapore','Dubai','US'],'Foreign'),inplace=True)
            
    dataframe['Loc'].replace({'North India' : 0,
                                'South India': 1,
                                'Foreign':2
                                       },inplace=True)
```

```python
encoding_demograph(data_before)
```

```python
#define a function for encoding features which follow a certain scale

def encode_scale3(dataframe):
    
    columns=dataframe.select_dtypes('object').columns
    
    for i in columns:
        if len(dataframe[i].value_counts()) == 3:
            dataframe[i].replace({'Disagree': 1,
                                        'Neutral':2,
                                        'Agree':3},inplace=True)
```

```python
encode_scale3(data_before)
```

```python
def encode_scale4(dataframe):
    
    columns=dataframe.select_dtypes('object').columns
    
    for i in columns:
        if len(dataframe[i].value_counts()) == 4:
            dataframe[i].replace({'Not interested' : 1,
                                         'Neutral': 2,
                                         'Moderately interested':3,
                                         'Very interested':4},inplace=True)
```

```python
encode_scale4(data_before)
```

```python
def encode_scale5(dataframe):
    
    columns=dataframe.select_dtypes('object').columns
    
    for i in columns:
        if len(dataframe[i].value_counts()) == 5:
            dataframe[i].replace({'Strongly disagree' : 1,
                                  'Disagree': 2,
                                  'Neutral':3,
                                  'Agree':4, 
                                  'Strongly Agree':5},inplace=True)
```

##### Before Data imputation using KNN and MICE

```python
from fancyimpute import KNN
from fancyimpute import IterativeImputer as MICE
```

```python
data_copy = data_before.copy()
data_before.drop('ID',axis=1,inplace=True)
```

```python
# use 5 nearest rows which have a feature to fill in each row's missing features
# fancy impute is going to remove the column names and convert the data into nummpy array, Hence added columns

data_before_final = pd.DataFrame(KNN(k=5).fit_transform(data_before),columns=data_before.columns)

# this will help us to replace the values
```

```python
### rounding off values
def roundvals(dataframe):
    for i in dataframe.columns:
        dataframe[i] = dataframe[i].apply(lambda x:round(x))
```

```python
roundvals(data_before_final)
```

```python
data_before_final['ID']=pd.Series(data_copy['ID'].values)
```

```python
# check for null values
data_before_final.isnull().sum()
```

```python
#15% of the data - 74 rows is being used as validation (2nd wave survey)

modeling_before=data_before_final.iloc[:447]

validation_before=data_before_final.iloc[447:]
```
